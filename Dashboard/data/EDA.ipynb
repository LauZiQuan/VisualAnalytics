```python
import pandas
import matplotlib.pyplot as plt
```

```python
hdb_property_information = pandas.read_csv('./hdb-property-information.csv')
```

```python
hdb_property_information['bldg_contract_town']
```

```python
def get_median_year(towns):
    for town in towns:
        current_town = hdb_property_information.loc[hdb_property_information['bldg_contract_town'] == town]
        print(town, current_town['year_completed'].median())
```

```python
def get_mode_year(towns):
    for town in towns:
        current_town = hdb_property_information.loc[hdb_property_information['bldg_contract_town'] == town]
        print(town, current_town['year_completed'].mode())
```

```python
def get_mean_year(towns):
    for town in towns:
        current_town = hdb_property_information.loc[hdb_property_information['bldg_contract_town'] == town]
        print(town, current_town['year_completed'].mean())
```
